##**demonstration of poisson distribution and gaussian distribution. **

also see http://individual.utoronto.ca/zheli/poisson.pdf  for a description of Poisson stochastic processes and a derivation of the Poisson distribution from the Bernoulli probability

```python
import os
import sys
import numpy as np
import pylab as pl
%pylab inline
```

#Poissont distribution: $\frac{\lambda^k}{k!} e^{-\lambda}$

```python
pl.hist(np.random.poisson(1,1000), bins=5, color='steelBlue', label=r'$\lambda=1$')
pl.legend()
pl.figure()
pl.hist(np.random.poisson(3,1000), color='IndianRed', label=r'$\lambda=3$')
pl.hist(np.random.poisson(1,1000), bins=5, color='steelBlue', alpha = 0.5, label=r'$\lambda=1$')
pl.legend()
pl.figure()
pl.hist(np.random.poisson(5,1000), color='DarkOrange', label=r'$\lambda=5$')
pl.hist(np.random.poisson(1,1000), bins=5, color='steelBlue', alpha = 0.5, label=r'$\lambda=1$')
pl.hist(np.random.poisson(3,1000), color='IndianRed', alpha = 0.5, label=r'$\lambda=3$')
pl.legend()
pl.figure()
pl.hist(np.random.poisson(10,1000), color='Green', label=r'$\lambda=10$')
pl.hist(np.random.poisson(1,1000), bins=5, color='steelBlue', alpha = 0.5, label=r'$\lambda=1$')
pl.hist(np.random.poisson(3,1000), color='IndianRed', alpha = 0.5, label=r'$\lambda=3$')
pl.hist(np.random.poisson(5,1000), color='DarkOrange', alpha = 0.5, label=r'$\lambda=5$')
pl.legend()

```

```python
#Gaussian istribution: $\frac{\lambda^k}{k!} e^{-\lambda}$
```

```python

pl.figure()
pl.hist(np.random.poisson(2,1000),color='Green', alpha = 0.5, label=r'Poisson $\lambda=1$')
pl.hist(np.random.normal(2,np.sqrt(2),1000), color='black', alpha = 0.5, label=r'Gaussian $\mu=1,\sigma=1$')
pl.plot([0,0],[0,300],'k-')
pl.legend()
pl.ylim(pl.ylim()[0],pl.ylim()[1]*1.2)
pl.figure()
pl.hist(np.random.poisson(10,1000), color='Green', alpha = 0.5, label=r'Poisson $\lambda=10$')
pl.hist(np.random.normal(10,np.sqrt(10),1000), color='black', alpha = 0.5, label=r'Gaussian $\mu=10,\sigma=\sqrt{10}$')
pl.legend()
pl.ylim(pl.ylim()[0],pl.ylim()[1]*1.2)
pl.figure()
pl.hist(np.random.poisson(1000,1000), color='Green', alpha = 0.5, label=r'Poisson $\lambda=10^3$')
pl.hist(np.random.normal(1000,np.sqrt(1000),1000), color='black', alpha = 0.5, label=r'Gaussian $\mu=10^3,\sigma=10^\frac{3}{2}$')
pl.legend()
pl.ylim(pl.ylim()[0],pl.ylim()[1]*1.2)

```
